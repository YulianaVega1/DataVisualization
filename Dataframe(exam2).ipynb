# **6. Filters and Pivots**

## Simple Filters

```python
# Filter by Elo and save it

challenger = dd.filter(like='Challenger',axis=0)
print(challenger)
```

```python
# Filter by Region and save it

lan = dd.filter(like='Lan',axis=0)
print(lan)
```

## Simple Pivots


```python
# Creating is a dataframe with the elo-region of summoners
er_sum = d.pivot(index = 'Elo',
                 columns = 'Region',
                 values = 'Summoners')
print(er_sum)

```

# **7. Basic line/Bar graph**

```python
# Line Graph the Elo of Lan

fig = px.line(challenger, x = challenger.index.get_level_values(0), y = 'Summoners')
# Change title
fig.update_layout(title = 'Elo of Lan',
                  title_font_size = 30,
                  title_x = 0.5)
# Change Axis
fig.update_xaxes(title = 'Region', title_font_size = 20)
fig.update_yaxes(title = 'Summoners', title_font_size = 20)

fig.show()
```

```python
# Bar Graph the Elo of Lan
fig = px.bar(challenger, x = challenger.index.get_level_values(0), y = 'Summoners')

# Change title
fig.update_layout(title = 'Elo of Lan',
                  title_font_size = 30,
                  title_x = 0.5,
                  yaxis_range = [50,400])

# Change Axis
fig.update_xaxes(title = 'Region', title_font_size = 20)
fig.update_yaxes(title = 'Summoners', title_font_size = 20)

fig.show()
```

## Two y-axis
